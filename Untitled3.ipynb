```python
#importer les librairies
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score,KFold,LeaveOneOut,ShuffleSplit
from sklearn import datasets, linear_model
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score,classification_report
```

```python
#lire le dataset
data=load_iris()
```

```python
#On met dans le variable X les attributs qui nous aidons Ã  expliquer le varibla y 
X=data['data']
```

```python
#le variable y c'est le variable qu'on veut predire
y=data['target']
```

```python
y
```

```python
x=data.data
y=data.target
plt.scatter(x[:,0],x[:,1],c=y,alpha=0.8)
```

# svm with traditionnel strategy

```python
#diviser le dataset en deux datasets train et test 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, alpha=0.8)
plt.title('Train set')
plt.subplot(122)
plt.scatter(X_test[:, 0], X_test[:, 1],c=y_test, alpha=0.8)
plt.title('Test set')
```

```python
#entrainement du model
classifier = svm.SVC(kernel='linear', random_state=None)
classifier.fit(X_train, y_train)
print('train score:', classifier.score(X_train, y_train))
print('test score:', classifier.score(X_test, y_test))
```

# svm with cross validation strategy

```python
cv=KFold(5)
cross_val_score(classifier,x,y,cv=cv).mean()
```

```python
cv=LeaveOneOut()
cross_val_score(classifier,x,y,cv=cv).mean()
```

```python
cv=ShuffleSplit(4,test_size=0.2)
cross_val_score(classifier,x,y,cv=cv).mean()
```

```python
classifier.score(X_test,y_test)
```

# appliquer le model sur les donneÃ©s du test

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
y_pred_vertical=y_pred.reshape(len(y_pred),1)
y_pred_vertical
```

```python
y_true_vertical=y_test.reshape(len(y_test),1)
y_true_vertical
```

```python
true_pred=np.concatenate((y_true_vertical,y_pred_vertical),axis=1)
true_pred
```

```python
matrix = classification_report(y_test,y_pred)
print('Rapport de Classification  : \n',matrix)
```

```python
def tracer_matrix_confusion(y_true,y_pred):
    import matplotlib.pyplot as plt
    from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
    cm = confusion_matrix(y_true,y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap=plt.cm.Blues)
    plt.suptitle("Matrice de confusion pour ")
    plt.show()
    return cm
```

```python
tracer_matrix_confusion(y_test,y_pred)
```

```python
print("F1-score pour chaque classe : \t",f1_score(y_test,y_pred,average=None))
print("taux de succÃ©es : \t\t",accuracy_score(y_test,y_pred))
print("rappel pour chaque classe : \t",recall_score(y_test,y_pred,average=None))
print("la precision pour chaque classe :",precision_score(y_test,y_pred,average=None))
print("taux d'erreurs : \t\t",1-accuracy_score(y_test,y_pred))
```
